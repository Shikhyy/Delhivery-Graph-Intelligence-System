# Delhivery Graph Intelligence - Part 3: ETA Prediction Modeling

This notebook demonstrates the core ML proposition: Graph-Enhanced ETA models outperform raw OSRM baselines.

```
import pandas as pd
import networkx as nx
from src.pipeline import load_and_clean, aggregate_to_trips
from src.graph_builder import build_graph
from src.graph_metrics import compute_node_metrics
from src.eta_model import train_baseline, train_graph_model, benchmark
from src.visualise import plot_model_comparison
```

```
df = load_and_clean('data/delivery_data.csv')
df_trip = aggregate_to_trips(df)
G = build_graph(df)
metrics = compute_node_metrics(G)
```

### 1. Train Baseline XGBoost
Using standard temporal, geospatial, and distance features.

```
baseline_results = train_baseline(df_trip)
```

### 2. Train Graph-Enhanced XGBoost
Injecting structural node2vec embeddings and bottleneck centrality scores.

```
graph_results = train_graph_model(df_trip, metrics, G)
```

### 3. Benchmark Comparison

```
bench_df = benchmark(baseline_results, graph_results)
plot_model_comparison(baseline_results, graph_results)
```
