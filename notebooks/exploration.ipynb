# Delhivery Graph Intelligence - Data Exploration

This notebook provides an interactive environment to explore the initial dataset and understand the logistics network distribution.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx

# Load the dataset
df = pd.read_csv("../data/delivery_data.csv")
df.head()
```

```
df.info()
```

```
# Missing values
df.isnull().sum()
```

```
# Route types distribution
sns.countplot(data=df, x="route_type")
plt.title("Distribution of Route Types (FTL vs Carting)")
plt.show()
```

```
# OSRM vs Actual Time deviation
df["delay_ratio"] = df["segment_actual_time"] / df["segment_osrm_time"]
plt.figure(figsize=(10, 5))
sns.histplot(df["delay_ratio"].clip(0, 3), bins=50, kde=True)
plt.axvline(1.0, color="r", linestyle="--", label="OSRM Baseline")
plt.axvline(1.2, color="orange", linestyle="--", label="20% Delay Threshold")
plt.title("Distribution of Segment Delay Ratios")
plt.legend()
plt.show()
```

```
# Exploring the hubs
top_sources = df["source_name"].value_counts().head(10)
top_sources.plot(kind="barh")
plt.title("Top 10 Source Hubs by Segment Volume")
plt.show()
```
